```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import os
import zipfile
from google.colab import drive

drive.mount('/content/drive')
```

```python
# Dosya adını 'archive.zip' olarak güncelledik
zip_yolu = '/content/drive/MyDrive/archive.zip'
cikarilacak_yer = '/content/dataset'

if not os.path.exists(cikarilacak_yer):
    with zipfile.ZipFile(zip_yolu, 'r') as zip_ref:
        zip_ref.extractall(cikarilacak_yer)
    print("Zıp dosyası başarıyla açıldı.")
else:
    print("Dosyalar zaten mevcut.")

# Kaggle'dan inen dosyalarda genelde içeriden 'chest_xray' klasörü çıkar.
# Bu yolları kontrol etmek için Colab sol menüdeki dosya simgesine bakabilirsin.
train_path = '/content/dataset/chest_xray/train'
test_path = '/content/dataset/chest_xray/test'
```

```python
# Eğitim için artırma, Test için sadece ölçeklendirme
train_gen = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_gen = ImageDataGenerator(rescale=1./255)

train_data = train_gen.flow_from_directory(train_path, target_size=(150, 150), batch_size=32, class_mode='binary')
test_data = test_gen.flow_from_directory(test_path, target_size=(150, 150), batch_size=32, class_mode='binary')
```

```python
model = Sequential()

# 1. Evrişim ve Havuzlama
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D(2, 2))

# 2. Evrişim ve Havuzlama
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(2, 2))

# 3. Evrişim (Hocanın en az 3 evrişim şartı için)
model.add(Conv2D(128, (3, 3), activation='relu'))

# Sınıflandırma Katmanları
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5)) # Düzenlileştirme şartı
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(train_data,
                    epochs=10,
                    validation_data=test_data)

# Grafikleri çizdir
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Eğitim')
plt.plot(history.history['val_accuracy'], label='Test')
plt.title('Doğruluk Grafiği')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Eğitim')
plt.plot(history.history['val_loss'], label='Test')
plt.title('Kayıp Grafiği')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Test verileri üzerinden tahmin yap
test_data.reset()
Y_pred = model.predict(test_data)
y_pred = (Y_pred > 0.5).astype(int)
y_true = test_data.classes

# Matrisi hesapla
cm = confusion_matrix(y_true, y_pred)

# Görselleştir
plt.figure(figsize=(6,6))
sns.heatmap(cm, annot=True, fmt='g', cmap='Blues',
            xticklabels=['Normal', 'Zaturre'],
            yticklabels=['Normal', 'Zaturre'])
plt.title('Şekil 4: Karışıklık Matrisi')
plt.ylabel('Gerçek Değer')
plt.xlabel('Tahmin Edilen')
plt.show()
```
